```python
r = requests.get('https://www.imdb.com/list/ls066620113/')
```

```python
r
```

```python
soup = bs4.BeautifulSoup(r.text)
```

```python
iron_man = 0
thor = 0
cpt_america = 0
for item in soup.find_all('div', class_ = 'list-description')[1:]:
    text = item.find('p').text
    if re.search('Tony Stark / Iron Man.+[\n]*', text) != None:
        print('Iron Man')
        subtext = re.search('Tony Stark / Iron Man.+[\n]*', text).group(0)[23:-2]
        print(subtext)
        print(subtext[:subtext.find(':')])
        print(subtext[subtext.find(':') + 1:])
        if subtext.find(':') == -1:
            iron_man += (int(subtext) * 60)
        else:
            if len(subtext[:subtext.find(':')]) == 0:
                iron_man += int(subtext[1:])
            else:
                iron_man += (int(subtext[:subtext.find(':')]) * 60)
                iron_man += int(subtext[subtext.find(':') + 1:])
    if re.search('Thor.+[\n]*', text) != None:
        print('Thor')
        subtext = re.search('Thor.+[\n]*', text).group(0)[6:-2]
        print(subtext)
        print(subtext[:subtext.find(':')])
        print(subtext[subtext.find(':') + 1:])
        if subtext.find(':') == -1:
            thor += (int(subtext) * 60)
        else:
            if len(subtext[:subtext.find(':')]) == 0:
                thor += int(subtext[1:])
            else:
                thor += (int(subtext[:subtext.find(':')]) * 60)
                thor += int(subtext[subtext.find(':') + 1:])
    if re.search('Steve Rogers / Captain America.+[\n]*', text):
        print('Captain America')
        subtext = re.search('Steve Rogers / Captain America.+[\n]*', text).group(0)[32:-2]
        print(subtext)
        print(subtext[:subtext.find(':')])
        print(subtext[subtext.find(':') + 1:])
        if subtext.find(':') == -1:
            cpt_america += (int(subtext) * 60)
        else:
            if len(subtext[:subtext.find(':')]) == 0:
                cpt_america += int(subtext[1:])
            else:
                cpt_america += (int(subtext[:subtext.find(':')]) * 60)
                cpt_america += int(subtext[subtext.find(':') + 1:])
```
